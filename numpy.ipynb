```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original Array:\n", arr)
print("Flattened Array:", arr.flatten())
```

```python
arr = np.array([1, 2, 3, 4])

copy_arr = arr.copy()
view_arr = arr.view()

arr[0] = 100

print("Original:", arr)
print("Copy:", copy_arr)   # Does NOT change
print("View:", view_arr)   # Changes because it shares memory
```

```python
arr = np.array([10, 15, 20, 25, 30, 35])

even = arr[arr % 2 == 0]
odd = arr[arr % 2 != 0]

print("Even Numbers:", even)
print("Odd Numbers:", odd)
```

```python
arr = np.array([5, 10, 15, 20, 25])

# Replace values greater than 15 with 100
arr[arr > 15] = 100

print("Modified Array:", arr)
```

```python
A = np.array([[4, -2],
              [1,  1]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```
