```python
# Power Calculation
num = 3
print("Power:", num ** 4)
```

```python
# Simple Interest
P = 1000
R = 5
T = 2
SI = (P * R * T) / 100
print("Simple Interest:", SI)
```

```python
#  String Upper and Lower
text = "python programming"
print("Upper:", text.upper())
print("Lower:", text.lower())

#  String Length and Count
print("Length:", len(text))
print("Count of 'p':", text.count('p'))

#  Replace Word
print("Replaced:", text.replace("python", "Java"))
```

```python
#  Append and Insert
nums = [1, 2, 3]
nums.append(4)
nums.insert(1, 10)
print("Updated List:", nums)

#  Sort and Reverse
nums = [5, 2, 9, 1]
nums.sort()
print("Sorted:", nums)
nums.reverse()
print("Reversed:", nums)

#  Pop Element
nums = [10, 20, 30, 40]
nums.pop()
print("After Pop:", nums)
```

```python
#  Count and Index
t = (1, 2, 3, 2, 4)
print("Count of 2:", t.count(2))
print("Index of 3:", t.index(3))

# Nested Tuple
nested = ((1,2), (3,4))
print("Nested Tuple:", nested)

#  Tuple Concatenation
t1 = (1,2)
t2 = (3,4)
print("Concatenated:", t1 + t2)
```

```python
#  Get Method
student = {"name": "Siva", "age": 21}
print("Get Age:", student.get("age"))

#  Remove Item
student.pop("age")
print("After Pop:", student)

#  Keys and Values
student = {"name": "Siva", "age": 21, "city": "Chennai"}
print("Keys:", student.keys())
print("Values:", student.values())
```
