```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

income = df["income"].to_numpy()
mean_income = np.mean(income)

plt.hist(income, bins=20)
plt.axvline(mean_income)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

x = df["income"].to_numpy()
y = df["sales"].to_numpy()

plt.scatter(x, y)
plt.title("Income vs Sales")
plt.xlabel("Income")
plt.ylabel("Sales")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

numeric_df = df.select_dtypes(include=[np.number])
corr = np.corrcoef(numeric_df.T)

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(numeric_df.columns)), numeric_df.columns, rotation=90)
plt.yticks(range(len(numeric_df.columns)), numeric_df.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

df["age_group"] = np.where(df["age"] < 30, "Young",
                     np.where(df["age"] < 50, "Middle", "Senior"))

grouped = df.groupby("age_group")["sales"].mean()

plt.bar(grouped.index, grouped.values)
plt.title("Average Sales by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Sales")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

grouped = df.groupby("gender")["sales"].sum()

plt.bar(grouped.index.astype(str), grouped.values)
plt.title("Total Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

x = df["income"].to_numpy()
y = df["sales"].to_numpy()

m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x + b)
plt.title("Linear Regression: Income vs Sales")
plt.xlabel("Income")
plt.ylabel("Sales")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

income = df["income"]
standardized = (income - np.mean(income)) / np.std(income)

plt.hist(standardized, bins=20)
plt.title("Standardized Income Distribution")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

df["age_group"] = np.where(df["age"] < 30, "Young",
                     np.where(df["age"] < 50, "Middle", "Senior"))

counts = df["age_group"].value_counts()

plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%')
plt.title("Age Group Distribution")
plt.show()
```
