```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    df[col + "_zscore"] = (df[col] - np.mean(df[col])) / np.std(df[col])

print(df.head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

df["risk_score"] = (0.4 * df["debt"] / np.max(df["debt"])) + \
                   (0.6 * df["miles"] / np.max(df["miles"]))

print(df[["debt", "miles", "risk_score"]].head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

numeric_df = df.select_dtypes(include=[np.number])

corr_matrix = np.corrcoef(numeric_df.T)

corr_df = pd.DataFrame(corr_matrix,
                       columns=numeric_df.columns,
                       index=numeric_df.columns)

print(corr_df)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

Q1 = np.percentile(df["income"], 25)
Q3 = np.percentile(df["income"], 75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["income"] < lower) | (df["income"] > upper)]

print("Outliers:\n", outliers)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

df["age_group"] = np.where(df["age"] < 30, "Young",
                     np.where(df["age"] < 50, "Middle", "Senior"))

print(df[["age", "age_group"]].head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

weights = np.array([0.2, 0.3, 0.5])

df["weighted_score"] = (df["age"] * weights[0] +
                        df["income"] * weights[1] +
                        df["sales"] * weights[2])

print(df[["age", "income", "sales", "weighted_score"]].head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

X = df[["age", "income", "debt"]].to_numpy()

coefficients = np.array([0.5, 0.3, 0.2])

result = np.dot(X, coefficients)

df["prediction_score"] = result

print(df[["age", "income", "debt", "prediction_score"]].head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    df[col + "_minmax"] = (df[col] - np.min(df[col])) / \
                          (np.max(df[col]) - np.min(df[col]))

print(df.head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

numeric_df = df.select_dtypes(include=[np.number])

cov_matrix = np.cov(numeric_df.T)

cov_df = pd.DataFrame(cov_matrix,
                      index=numeric_df.columns,
                      columns=numeric_df.columns)

print(cov_df)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

df["log_income"] = np.log(df["income"])

print(df[["income", "log_income"]].head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

sales_array = df["sales"].to_numpy()

ranks = np.argsort(np.argsort(-sales_array)) + 1

df["sales_rank"] = ranks

print(df[["sales", "sales_rank"]].head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")

threshold = np.mean(df["income"])

df["high_income"] = np.where(df["income"] > threshold, 1, 0)

print(df[["income", "high_income"]].head())
```
